```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.max_columns', None)
filepath = 'Datasets/Kaggle/Life Expectancy Kaggle.csv'
df = pd.read_csv(filepath)
df
```

```python
df.dtypes
```

```python
df.columns = ['Country', 'Year', 'Status', 'Life expectancy', 'Adult mortality', 'Infant deaths', 'Alcohol', 'Percentage expenditure', 'Hepatitis B',
             'Measles', 'BMI', 'Under-five deaths', 'Polio', 'Total expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'Thinness 1-19 years', 
              'Thinness 5-9 years', 'Income composition of resources', 'Schooling']
df.drop(columns='Status', axis=1, inplace=True)
df
```

```python
df['Country'].unique()
```

```python
print(df[['Country', 'Year']].nunique())
```

```python
nan_colcounts = df.isna().sum()/df.shape[0]*100
nan_colcounts
```

```python
df = df.dropna(subset=['Life expectancy'])
df.reset_index(drop=True, inplace=True)
df
```

```python
nan_colcounts = df.isna().sum()/df.shape[0]*100
nan_colcounts
```

```python
nan_yearcounts = df.isna().groupby(df['Year']).sum()
plt.figure(figsize=(10, 6))
nan_yearcounts.sum(axis=1).plot(kind='bar')  # Summing NaN counts across all columns for each year
plt.title('Count of NaN Values Per Year')
plt.xlabel('Year')
plt.ylabel('Number of NaN Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df = df[df['Year'] < 2015]
df
```

```python
nan_colcounts = df.isna().sum()/df.shape[0]*100
nan_colcounts
```

```python
nan_rowcounts = df.isna().sum(axis=1)/df.shape[1] * 100
plt.figure(figsize=(10, 6))
plt.hist(nan_rowcounts, bins=list(range(0, 101, 5)), color='skyblue', edgecolor='black')
plt.title('Histogram of Missing Value Percentages per Row')
plt.xlabel('Percentage of Missing Values')
plt.ylabel('Number of Rows')
plt.grid(True)
plt.show()
```

```python
df = df[nan_rowcounts < 25].copy()
df
```

```python
df.sort_values(by=['Country', 'Year'], inplace=True)

exclude_columns = ['Country', 'Year', 'Status', 'Life expectancy']

for attribute in df.columns:
    if attribute not in exclude_columns:
        df.loc[:,attribute] = df.groupby('Country')[attribute].transform(lambda group: group.interpolate())


nan_colcounts = df.isna().sum()/df.shape[0] * 100
print(nan_colcounts)
```

```python
for attribute in df.columns:
    if attribute not in exclude_columns:
        # Use direct methods ffill() and bfill() and ensure the index aligns with the DataFrame
        df.loc[:,attribute] = df.groupby('Country')[attribute].transform(
            lambda group: group.ffill().bfill()
        )
nan_colcounts = df.isna().sum()/df.shape[0] * 100
print(nan_colcounts)
```

```python
number_of_countries = df['Country'].nunique()
number_of_countries
```

```python
attribute_name = 'GDP' 

countries_no_data = df.groupby('Country')[attribute_name].apply(lambda x: x.isna().all())

countries_no_data = countries_no_data[countries_no_data]

countries_no_data.shape[0] / number_of_countries * 100
```

```python
attribute_name = 'Population' 

countries_no_data = df.groupby('Country')[attribute_name].apply(lambda x: x.isna().all())

countries_no_data = countries_no_data[countries_no_data]

countries_no_data.shape[0] / number_of_countries * 100
```

```python
df = df.loc[:,nan_colcounts < 10].copy()
df
```

```python
print(df.isna().any(axis=1).sum())
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
nan_colcounts = df.isna().sum()/df.shape[0] * 100
print(nan_colcounts)
```
